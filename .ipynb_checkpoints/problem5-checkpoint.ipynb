```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sps
```

```python
with np.load('data_interp.npz') as data:
    x1, y1 = data['x1'], data['y1']
```

```python
from scipy.interpolate import lagrange
from scipy.interpolate import CubicSpline
from scipy.interpolate import PchipInterpolator
```

```python
plt.figure(figsize=(8, 6))
plt.subplot(2, 2, 1)
plt.plot(x1, y1, label='original')
poly = lagrange(x1, y1)
plt.plot(x1, poly(x1), label='lagrange', linestyle='--')
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(x1, y1, label='original')
cs = CubicSpline(x1, y1)
plt.plot(x1, cs(x1), label='CubicSpline', linestyle='--')
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(x1, y1, label='original')
pc = PchipInterpolator(x1, y1)
plt.plot(x1, pc(x1), label='PchipInterpolator', linestyle='--')
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(x1, y1, label='original')
coeff, [err], *_ =  np.polyfit(x1, y1, 17, full=True)
plt.plot(x1, np.polyval(coeff,x1), label='Poly', linestyle='--')
plt.legend()

plt.show()
```

```python
from numpy.polynomial.polynomial import Polynomial
```

```python
n = 100
x = np.linspace(-2, 2, n)
y = 1/(1 + 25*x**2)
plt.figure(figsize=(8, 6))
plt.plot(x, y, '-', label = 'original')
m = [3, 5, 7, 11]  #   
xo = []
yo = []
polyt = []
coefs = []
yj = []
for i in range(len(m)):
    xo.append(np.arange(-2, 2.1, 4/m[i]))
    yo.append(np.array([]))
    yo[i] = 1/(1 + 25*xo[i]**2)
    polyt.append(lagrange(xo[i], yo[i]))
    coefs.append(Polynomial(polyt[i].coef[::-1]).coef)
    #print(polyt[i])
    
for i in range(len(m)):
    yj.append(np.zeros(n))
    for j in range(len(coefs[i])):
        yj[i] += coefs[i][j] * x ** (j)

plt.plot(x, yj[3], '--', label = '$x^{11}$')
plt.plot(x, yj[2], '--', label = '$x^{7}$')
plt.plot(x, yj[1], '--', label = '$x^{5}$')
plt.plot(x, yj[0], '--', label = '$x^{3}$')
plt.legend()
plt.grid()
```

```python
import numpy.polynomial.chebyshev as cheb
def cheb_nodes(n, a=-1, b=1):
    r"""Chebyshev nodes of degree $n$ on $[a, b]$
    """
    nodes = np.array([])
    for i in range(1, n + 1):
      nodes = np.append(nodes, 1/2*(a+b)+1/2*(b-a)*np.cos((2*i-1)*np.pi/(2*n)))
    return sorted(nodes)
```

```python
nodes_11 = cheb_nodes(11)
nodes_11 = np.asarray(nodes_11)
assert (nodes_11[1:] > nodes_11[:-1]).all()

from scipy.special import roots_chebyt
nodes, weights = roots_chebyt(5) 
```

```python
x = np.arange(-2, 2.1, 0.02)
y = 1 / (1 + 25 * x ** 2)

def newdata(n):
    x1 = np.array(cheb_nodes(n, -2, 2))
    y1 = 1 / (1 + 25 * x1 ** 2)
    return x1, y1

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='original')
poly = lagrange(newdata(3)[0], newdata(3)[1])
plt.plot(x, poly(x), label='lagrange m=3',linestyle='-')

poly = lagrange(newdata(5)[0], newdata(5)[1])
plt.plot(x, poly(x), label='lagrange m=5',linestyle='-')

poly = lagrange(newdata(7)[0], newdata(7)[1])
plt.plot(x, poly(x), label='lagrange m=7',linestyle='-')

poly = lagrange(newdata(11)[0], newdata(11)[1])
plt.plot(x, poly(x), label='lagrange m=11',linestyle='-')

cs = CubicSpline(newdata(3)[0], newdata(3)[1])
plt.plot(x, cs(x), label='cub 3',linestyle='-.')

cs = CubicSpline(newdata(5)[0], newdata(5)[1])
plt.plot(x, cs(x), label='cub 5',linestyle='-.')

cs = CubicSpline(newdata(7)[0], newdata(7)[1])
plt.plot(x, cs(x), label='cub 7',linestyle='-.')

cs = CubicSpline(newdata(11)[0], newdata(11)[1])
plt.plot(x, cs(x), label='cub 11',linestyle='-.')


plt.legend()
plt.grid()
```
